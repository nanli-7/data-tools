```python
import urllib
from bs4 import BeautifulSoup
```

```python
page = urllib.request.urlopen("https://simple.wikipedia.org/wiki/Artificial_intelligence").read()
soup = BeautifulSoup(page,"html.parser")
```

```python
soup.title
```

```python
soup.p
```

```python
# extracting content found within a page’s <p> tags:
for elem in soup.find_all('p'):
    print(elem.text)
```

```python
ai_wiki = ''
for elem in soup.find_all('p'):
    ai_wiki += elem.text

```

```python
ai_wiki
```

```python
# save string as a text file

with open("ai-wiki.txt", "w") as text_file:
    text_file.write(ai_wiki)
```
