```python
import pandas
```

```python
import numpy
```

```python
df=pandas.read_csv('../data/avocado.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
pandas.pivot_table(df, values=['AveragePrice','Total Volume'], index=['Date', 'region'],
                   columns=['type'])
```

```python
pandas.pivot_table(df, values=['AveragePrice','Total Volume'], index=[ 'region'],
                   columns=['type'],  aggfunc={'AveragePrice': numpy.mean,
                                               'Total Volume': numpy.mean})
```

### average price and total sale volume for each year

```python
pandas.pivot_table(df, values=['AveragePrice','Total Volume'], 
                   columns= 'type',
                   index=['region', 'year'],
                   aggfunc={'AveragePrice': numpy.mean,
                            'Total Volume': numpy.sum})
```

```python
TotalUS = df[df['region'] == 'TotalUS']
```

```python
pandas.pivot_table(TotalUS, values=['AveragePrice','Total Volume'], 
                   columns= 'type',
                   index=['region', 'year'],
                   aggfunc={'AveragePrice': [min,max],
                            'Total Volume': numpy.sum})
```
