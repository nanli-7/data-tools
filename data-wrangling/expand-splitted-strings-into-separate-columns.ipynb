```python
import pandas
```

```python
ebola = pandas.read_csv('../data/country_timeseries.csv')
```

```python
ebola.head()
```

```python
ebola_long = pandas.melt(ebola, id_vars=['Date', 'Day'])
```

```python
ebola_long.head()
```

### 1. Setup expand parameter

```python
variable_split = ebola_long['variable'].str.split('_', expand=True)
```

```python
variable_split.head()
```

```python
type(variable_split)
```

```python
variable_split.columns = ['status_expand', 'country_expand']
```

```python
ebola_long = pandas.concat([ebola_long, variable_split], axis=1)
```

```python
ebola_long.head()
```

### 2. Simpler way

```python
ebola_long[['status_sim', 'country_sim']] = ebola_long['variable'].str.split('_', expand=True)
```
