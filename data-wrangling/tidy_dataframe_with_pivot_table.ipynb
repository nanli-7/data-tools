```python
import pandas
```

```python
weather = pandas.read_csv('../data/weather.csv')
```

```python
weather.head()
```

```python
weather.shape
```

```python
weather_melt = pandas.melt(
    weather,
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt.head()
```

```python
#dropna: boolean, default True, will drop NaN value

weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```
