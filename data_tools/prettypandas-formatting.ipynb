__PrettyPandas__ is a library that provides an easy way to format data-frames and to add table summaries to them.

**Features**

* Add summary rows and columns.
* Number formatting for currency, scientific units, and percentages.
* Chaining commands.
* Works seamlessly with Pandas Style API.

```python
import pandas 
import numpy
```

```python
import prettypandas
```

```python
df = pandas.DataFrame(3 * numpy.random.random(size=(10, 5)), columns=list('ABCDE'))
```

```python
df\
.pipe(prettypandas.PrettyPandas)\
.as_currency('EUR', subset='A')\
.total()\
.average()
```
